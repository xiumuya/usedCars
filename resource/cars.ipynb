```python
# 按照上个单元格的wordM 替换对应unicode编码（在原始json文件里）
import json
import re
wordMap = {}
with open('wordM.json', 'r') as file:
    wordMap = json.load(file)

# 1. 读取JSON文件
for jsonPath in json_files:
    with open(jsonPath, 'r') as file:
        content = json.load(file)

    # 2. 将字典转换为字符串
    json_str = json.dumps(content)

    # 3. 使用re.sub按键值对进行替换
    for key, value in wordMap.items():
        # 注意：使用re.escape来处理特殊字符
        json_str = re.sub(re.escape(key), value, json_str)

    # 4. 将字符串转换回字典
    updated_data = json.loads(json_str)

    # 5. 将更新后的字典保存回JSON文件
    with open(r".\json_replace"+jsonPath.split("\\json")[1], 'w') as file:
        json.dump(updated_data, file, indent=4)
```

```python
import requests
from fake_useragent import UserAgent
from time import sleep
from lxml import etree
from tqdm import tqdm

# 用户代理生成
ua = UserAgent()
user_agent = ua.random

# 添加请求头
headers = {
    "User-Agent": user_agent,
    "Cookie": r"tt_web_version=new; ttwid=1%7CpgEw87Bg6p7K5mYiEXAgCKqcGS4LYw6FO0t0jCvpl6Y%7C1731905217%7C882d731ee7e34c9a47cd43459eda519ea4e76a650e16bb0355555326dd1d07e2; tt_webid=7438476096486016536; _gid=GA1.2.744737134.1731905217; city_name=%E9%87%8D%E5%BA%86; s_v_web_id=verify_m3mjnf7b_OBOF5UhI_Rudx_4ch7_AdBz_wRLg1EqXDkUc; rit_city=%E9%87%8D%E5%BA%86; is_dev=false; is_boe=false; Hm_lvt_3e79ab9e4da287b5752d8048743b95e6=1731905916,1731910307,1731911430; HMACCOUNT=8160C1392AF61C82; Hm_lpvt_3e79ab9e4da287b5752d8048743b95e6=1731911723; _ga_YB3EWSDTGF=GS1.1.1731908420.13.1.1731914078.60.0.0; _ga=GA1.1.427450835.1721306153",
}

# 初始化列表
car_level = []
car_type_of_energy = []
car_maximum_speed = []
car_100_meters_acceleration_time = []
car_door_num = []
car_seat_num = []

# 提取数据的函数
def extract_data(html, label_text):
    try:
        # 使用 XPath 定位元素
        result = html.xpath(f'//label[text()="{label_text}"]/ancestor::div[@class="table_col__3Pc3_ table_is-sticky__1hkPu table_is-label__1wIhd"]/following-sibling::div//div[@class="cell_normal__37nRi"]/text()')
        # 如果结果为空，返回 "未知"
        return result[0] if result else "未知"
    except Exception as e:
        print(f"Error extracting {label_text}: {e}")
        return "未知"

# 假设 cars_duplicated 是包含 car_id 的 DataFrame
carsID = cars_duplicated["car_id"]

# 循环通过 carsID 获取数据
for i in tqdm(carsID[800:809]):
    # 构建请求 URL
    href = f"https://www.dongchedi.com/auto/params-carIds-{i}"

    # 发送请求
    res = requests.get(url=href, headers=headers)

    # 确保请求成功
    res.raise_for_status()  # 如果请求失败会抛出异常

    # 解析返回的 HTML 内容
    res.encoding = res.apparent_encoding  # 设置编码
    try:
        html = etree.HTML(res.text)  # 重新解析每次请求的 HTML
        if html is None:
            print(f"Error: Failed to parse HTML for car ID {i}")
            continue
    except Exception as e:
        print(f"Error parsing HTML for car ID {i}: {e}")
        continue

    # 提取数据并追加到对应的列表
    car_level.append(extract_data(html, "级别"))
    car_type_of_energy.append(extract_data(html, "能源类型"))
    car_maximum_speed.append(extract_data(html, "最高车速(km/h)"))
    car_100_meters_acceleration_time.append(extract_data(html, "官方百公里加速时间(s)"))
    car_door_num.append(extract_data(html, "车门数(个)"))
    car_seat_num.append(extract_data(html, "座位数(个)"))

    # 暂停 0.2 秒，避免过于频繁的请求
    sleep(0.2)

# 输出结果
print(f"车级别: {car_level}")
print(f"能源类型: {car_type_of_energy}")
print(f"最高车速: {car_maximum_speed}")
print(f"百公里加速时间: {car_100_meters_acceleration_time}")
print(f"车门数: {car_door_num}")
print(f"座位数: {car_seat_num}")
```
